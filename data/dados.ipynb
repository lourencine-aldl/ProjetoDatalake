```python
import pandas as pd
import os

# Caminho da pasta de dados
data_folder = '/home/Projetos/ProjetosDocker/data'

# Listar todos os arquivos CSV
csv_files = [f for f in os.listdir(data_folder) if f.endswith('.csv')]
csv_files.sort()

print(f'Encontrados {len(csv_files)} arquivos CSV:')
for f in csv_files:
    print(f'  - {f}')
```

```python
# Ler todos os arquivos CSV em um dicionário
dataframes = {}

for csv_file in csv_files:
    file_path = os.path.join(data_folder, csv_file)
    try:
        df = pd.read_csv(file_path)
        dataframes[csv_file] = df
        print(f'\n{csv_file}:')
        print(f'  Shape: {df.shape}')
        print(f'  Colunas: {list(df.columns)}')
    except Exception as e:
        print(f'Erro ao ler {csv_file}: {e}')
```

```python
# Ler e explorar o arquivo data-1772046577622.csv
df = dataframes['data-1772046577622.csv']

print('=== INFORMAÇÕES DO ARQUIVO ===')
print(f'Total de linhas: {df.shape[0]}')
print(f'Total de colunas: {df.shape[1]}')
print(f'\nTipos de dados:')
print(df.dtypes)
print(f'\n=== PRIMEIRAS 5 LINHAS ===')
df.head()
```

```python
# Estatísticas e valores ausentes
print('=== VALORES AUSENTES ===')
print(df.isnull().sum())

print('\n=== ESTATÍSTICAS DAS COLUNAS NUMÉRICAS ===')
print(df.describe())

print('\n=== VALORES ÚNICOS POR COLUNA ===')
for col in df.select_dtypes(include='object').columns:
    unique_count = df[col].nunique()
    if unique_count <= 10:
        print(f'{col}: {unique_count} valores únicos')
        print(f'  Valores: {df[col].unique()[:10].tolist()}')
    else:
        print(f'{col}: {unique_count} valores únicos')
```

```python
# Análise detalhada: Distribuição de Status e Fonte
print('=== DISTRIBUIÇÃO DE STATUS ===')
print(df['status'].value_counts())

print('\n=== DISTRIBUIÇÃO DE FONTE ===')
print(df['source'].value_counts())

print('\n=== DISTRIBUIÇÃO DE RAZÃO DE MATCH ===')
print(df['match_reason'].value_counts())

print('\n=== REGISTROS POR RUN_ID ===')
print(df['run_id'].value_counts())
```

```python
# Top 10 Suppliers com mais registros
print('=== TOP 10 SUPPLIERS ===')
print(df['supplier'].value_counts().head(10))

print('\n=== ESTATÍSTICAS DE AMOUNT POR STATUS ===')
print(df.groupby('status')['amount'].agg(['count', 'sum', 'mean', 'min', 'max']))

print('\n=== DADOS CONSOLIDADOS ===')
print(f'Total de registros: {len(df)}')
print(f'Registros sem match: {len(df[df["status"] == "UNMATCHED"]) + len(df[df["status"] == "UNMATCHED_MOVEMENT"])}')
print(f'Taxa de match: {len(df[df["status"] == "MATCHED"]) / len(df) * 100:.2f}%')
print(f'Valor total: R$ {df["amount"].sum():,.2f}')
```
